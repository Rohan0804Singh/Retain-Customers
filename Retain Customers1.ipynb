**5)What is the distribution of subscription types?**


```python
#Creating a function to find the count of people taking the service and who are not
def count1(column):
    count1=0
    count2=0
    for i in column:
        if i=='Yes':
            count1=count1+1
        else:
            count2=count2+1
    print("Yes-",count1)
    print("No-",count2)

        
```

```python
#For Phone Service
count1(df['PhoneService'])
```

```python
#For MulipleLines
count1(df['MultipleLines'])
```

```python
#For online security
count1(df['OnlineSecurity'])
```

```python
#For Online Backup
count1(df['OnlineBackup'])
```

```python
#Creating a new function for internet service column
def internet(column):
    count1=0
    count2=0
    for i in column:
        if i=='DSL' or i=='Fiber optic':
            count1=count1+1
        else:
            count2=count2+1
    print('Yes-',count1)
    print('No-',count2)
            
    
```

# Conclusion and Insights To Retain Customers

* **Since the churn rate of senior citizens is high as compared to other age groups so service providers should focus more to provide services more suitable for senior citizens. It can be done by including discounts or addtitional benefits or adding a service that is more suitable for senior citiens**.
* **Majority of people who churn have a month to month subscription type. So the service provider can add some additional benefits for the next month for the customer to avoid churning.**.
* **Apart from Phone Services and Internet Service other services have very less subcriptions,so the service providers should focus on these services to increase the subscriptions**.
* **Service providers should focus more on the services which cost from 0 to 2000 since majority of customers pay total charges in this range**.
* **Services are equally distributed in male and females,i.e,churn does not vary across gender**.
